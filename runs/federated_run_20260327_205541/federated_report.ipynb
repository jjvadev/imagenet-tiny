```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

run_dir = Path(r"runs\federated_run_20260327_205541")
with open(run_dir / "metrics.json", "r", encoding="utf-8") as f:
    history = json.load(f)
df = pd.DataFrame(history)
df
```

```python
print('Última ronda:')
display(df.tail(1))
print('Mejor val_acc:')
display(df.loc[[df['val_acc'].idxmax()]])
print('Mejor val_top5:')
display(df.loc[[df['val_top5'].idxmax()]])
print('Menor val_loss:')
display(df.loc[[df['val_loss'].idxmin()]])
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['round'], df['train_loss'], marker='o', label='Train Loss')
plt.plot(df['round'], df['val_loss'], marker='o', label='Val Loss')
plt.title('Loss per Round')
plt.xlabel('Round')
plt.ylabel('Loss')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['round'], df['train_acc'], marker='o', label='Train Acc')
plt.plot(df['round'], df['val_acc'], marker='o', label='Val Acc')
plt.title('Top-1 Accuracy per Round')
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['round'], df['train_top5'], marker='o', label='Train Top5')
plt.plot(df['round'], df['val_top5'], marker='o', label='Val Top5')
plt.title('Top-5 Accuracy per Round')
plt.xlabel('Round')
plt.ylabel('Accuracy')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
