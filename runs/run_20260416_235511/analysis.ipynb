```python
plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['cost'], marker='o')
plt.title('Cost por epoch')
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['train'], marker='o', label='Train')
plt.plot(df['epoch'], df['test'], marker='s', label='Test')
plt.title('Accuracy por epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(df['epoch'], df['epoch_time'], marker='o', label='Tiempo por epoch')
plt.plot(df['epoch'], df['total_time'], marker='s', label='Tiempo acumulado')
plt.title('Tiempo de entrenamiento')
plt.xlabel('Epoch')
plt.ylabel('Segundos')
plt.grid(True)
plt.legend()
plt.show()
```
